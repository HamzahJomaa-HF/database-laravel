```python
# -------------------------------
# Jupyter/Python Dummy Data Script
# -------------------------------

import uuid
from datetime import datetime, timedelta
from sqlalchemy import create_engine, text
from faker import Faker
from dotenv import load_dotenv
import os

# -------------------------------
# 1️⃣ Load environment variables
# Make sure .env is in the same folder as this notebook
# -------------------------------
load_dotenv()

db_user = os.getenv("DB_USERNAME")
db_password = os.getenv("DB_PASSWORD")
db_host = os.getenv("DB_HOST")
db_port = os.getenv("DB_PORT")
db_name = os.getenv("DB_DATABASE")

# -------------------------------
# 2️⃣ Connect to PostgreSQL
# -------------------------------
engine = create_engine(
    f'postgresql+psycopg2://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}'
)

# -------------------------------
# 3️⃣ Faker instance
# -------------------------------
fake = Faker()

# -------------------------------
# 4️⃣ Insert dummy programs
# -------------------------------
programs = []

with engine.begin() as conn:  # start a transaction
    for _ in range(5):
        program_id = str(uuid.uuid4())
        name = fake.company().replace("'", "")
        description = fake.text(max_nb_chars=50).replace("'", "")
        created_at = updated_at = datetime.now()

        conn.execute(
            text(
                f"INSERT INTO programs (program_id, name, description, created_at, updated_at) "
                f"VALUES ('{program_id}', '{name}', '{description}', '{created_at}', '{updated_at}')"
            )
        )
        programs.append(program_id)

print("✅ Dummy programs inserted.")

# -------------------------------
# 5️⃣ Insert dummy project centers (safe with parent IDs)
# -------------------------------
project_centers = []

with engine.begin() as conn:
    # Step 1: Insert all project centers without parents
    for _ in range(5):
        project_center_id = str(uuid.uuid4())
        program_id = fake.random_element(elements=programs)
        project_type = fake.word()
        project_group = fake.company().replace("'", "")
        external_id = str(uuid.uuid4())
        created_at = updated_at = datetime.now()

        conn.execute(
            text(
                f"INSERT INTO project_centers "
                f"(project_center_id, program_id, project_type, project_group, external_id, created_at, updated_at) "
                f"VALUES "
                f"('{project_center_id}', '{program_id}', '{project_type}', '{project_group}', '{external_id}', '{created_at}', '{updated_at}')"
            )
        )
        project_centers.append(project_center_id)

    print("✅ Step 1 done: project centers inserted without parents.")

    # Step 2: Assign random parent IDs (must exist already)
    for pc_id in project_centers:
        parent_id = fake.random_element(elements=project_centers)
        # Make sure a record doesn't become its own parent
        if parent_id == pc_id:
            continue

        conn.execute(
            text(
                f"UPDATE project_centers "
                f"SET parent_project_center_id = '{parent_id}' "
                f"WHERE project_center_id = '{pc_id}'"
            )
        )

print("✅ Step 2 done: parent_project_center_id assigned.")


# -------------------------------
# 6️⃣ Insert dummy activities
# -------------------------------
with engine.begin() as conn:
    for _ in range(10):
        activity_id = str(uuid.uuid4())
        activity_title = fake.bs().replace("'", "")
        activity_type = fake.word()
        content_network = fake.text(max_nb_chars=100).replace("'", "")
        start_date = fake.date_between(start_date='-1y', end_date='today')
        end_date = start_date + timedelta(days=fake.random_int(min=1, max=30))
        external_id = str(uuid.uuid4())
        target_cop = None
        parent_activity = None
        created_at = updated_at = datetime.now()

        conn.execute(
            text(
                f"INSERT INTO activities "
                f"(activity_id, activity_title, activity_type, content_network, start_date, end_date, external_id, target_cop, parent_activity, created_at, updated_at) "
                f"VALUES "
                f"('{activity_id}', '{activity_title}', '{activity_type}', '{content_network}', '{start_date}', '{end_date}', '{external_id}', {f'\'{target_cop}\'' if target_cop else 'NULL'}, {f'\'{parent_activity}\'' if parent_activity else 'NULL'}, '{created_at}', '{updated_at}')"
            )
        )

print("✅ Dummy activities inserted.")
print("🎉 All dummy data inserted successfully!")
# -------------------------------
# 7️⃣ Insert dummy users
# -------------------------------
with engine.begin() as conn:
    for _ in range(10):
        user_id = str(uuid.uuid4())
        first_name = fake.first_name().replace("'", "")
        last_name = fake.last_name().replace("'", "")
        dob = fake.date_of_birth(minimum_age=18, maximum_age=65)
        phone_number = fake.phone_number().replace("'", "")
        created_at = updated_at = datetime.now()

        conn.execute(
            text(
                f"INSERT INTO users "
                f"(user_id, first_name, last_name, dob, phone_number, created_at, updated_at) "
                f"VALUES "
                f"('{user_id}', '{first_name}', '{last_name}', '{dob}', '{phone_number}', '{created_at}', '{updated_at}')"
            )
        )

print("✅ Dummy users inserted.")
```
